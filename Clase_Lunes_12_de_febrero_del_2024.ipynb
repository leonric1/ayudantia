```python
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

x= np.linspace(-8,8,500)
y= np.linspace(-8,8,500)

X,Y = np.meshgrid(x,y)

Z = X**2/25 + Y**2/16

plt.contour (X,Y,Z, levels=[1], colors='red')
plt.xlabel('x')
plt.xlabel('y')
plt.title('Elipse centrada en el origen')
plt.show()
```
